# 2. Chargement des données
## 2.1 Import des librairies

````python
pip install pandas
pip install pyroj
````

````python
import pandas as pd
from pyproj import Transformer
import os
````

## 2.2 Chargement du fichier CSV

````python
# 1.Chargement du dataset

base_dir = os.path.dirname(os.path.abspath(__file__))
chemin = os.path.join(base_dir, "referentiel-archeologique-de-paris.csv")

df = pd.read_csv(chemin, sep=";", encoding="utf-8")

df.head()
````

## 2.3 Nettoyage du dataset

````python
# 2.Suppression des colonnes inutiles (Commune, colonnes de périodes ("Antiquité", "Moyen-Âge"...) qui sont tout le temps vides).
# -> Le dataset ne comprends que des chantiers ayant eu lieu à Paris, inutile de garder cette colonne qui ne contient que la valeur "Paris".

colonnes_inutiles = [
    "Commune",
    "Préhistoire",
    "Protohistoire",
    "Antiquité",
    "Moyen-Age",
    "Temps modernes",
    "Epoque contemporaine"
]

df = df.drop(columns=colonnes_inutiles)
````

````python
# 3. Renommage des colonnes.
# -> La base de données a pour nouvelle table principale "descriptifs_chantiers". Le dataset ne comporte qu'une liste de chantiers de fouilles.
# -> Toutes les données et noms de colonnes sont entièrement en français.

df = df.rename(columns={
    "Identifiant": "id_chantier",
    "Nature de l'opération": "nature_operation",
    "Responsable de l'opération": "responsable",
    "Date de l'opération": "date_operation",
    "Synthèse": "synthese"
})
````

````python
# 4. Nouveau format des dates.
# -> PostgreSQL a été retenu pour la migration. Format de date : AAAA--MM-JJ (YYYY-MM-DD).

df["date_operation"] = pd.to_datetime(df["date_operation"], errors="coerce")
````

````python
# 5.Nettoyage des colonnes : TEXT

df["nature_operation"] = df["nature_operation"].str.strip().str.lower()
df["responsable"] = df["responsable"].fillna("").str.strip().str.title()
df["synthese"] = df["synthese"].str.strip()
````

````python
# 6. Colonne période.
# -> Plutôt que multiples colonnes par période (qui sont vides dans le CSV), utiliser une colonne unique (vide au départ).
# -> La colonne pourra être remplie ultérieurement.

df["periode"] = None
````

````python
###### CREATION DES TABLES SUPPLEMENTAIRES #######

# 7. Table "Localisation".
# -> Les colonnes latitude/longitude s'appellement originellement dans le CSV X/Y. A renommer.

transformer = Transformer.from_crs("EPSG:2154", "EPSG:4326", always_xy=True)
df["longitude"], df["latitude"] = transformer.transform(df["X"], df["Y"])

# "id_chantier" reprends les valeurs de la colonne "Identifiant".

localisation = df[[
    "id_chantier",
    "Adresse",
    "Code postal",
    "latitude",
    "longitude"
]].copy()

localisation = localisation.rename(columns={
    "Adresse": "adresse",
    "Code postal": "code_postal"
})
````

````python
###### Export final ###

df.to_csv(os.path.join(base_dir, "descriptifs_chantiers_clean.csv"), index=False)
localisation.to_csv(os.path.join(base_dir, "localisation_clean.csv"), index=False)
````

# <span style="color:#2F3E46;">4. Analyse de qualité des données</span>

---

## <span style="color:#2F3E46;">4.1 Valeurs manquantes</span>

L’examen du dataset brut révèle que les colonnes dédiées aux périodes historiques  
(<i>Antiquité</i>, <i>Moyen Âge</i>, <i>Moderne</i>, <i>Contemporaine</i>) étaient entièrement vides.  
Lors du nettoyage, ces champs ont été fusionnés en une seule colonne <b>période</b> afin de simplifier la structure.

Cependant, l’absence totale d’information structurée sur les périodes constitue un manque important :  
les indications chronologiques sont en réalité <b>dissimulées dans les textes descriptifs</b> de la colonne <i>synthèse</i>.

<div style="border-left: 6px solid #8A5A44; padding: 12px; background-color:#F4F1EC; border-radius: 6px;">
<b>Interprétation :</b> Cette situation complique l’analyse et révèle une fragilité documentaire : un champ essentiel en archéologie n’a pas été renseigné dans le fichier source.
</div>

---

## <span style="color:#2F3E46;">4.2 Doublons</span>

Quelques doublons apparents ont été identifiés, notamment des occurrences répétées d’adresses comme  
<b>Boulevard Saint‑Marcel</b>.  
Ces répétitions ne correspondent pas à de véritables doublons : les numéros diffèrent, ce qui indique des interventions distinctes le long d’un même axe urbain.

De même, plusieurs opérations datées de la fin du XIXᵉ siècle sont associées à <b>Théodore Vacquier</b>, figure majeure de l’archéologie parisienne.  
Ces occurrences multiples reflètent probablement des interventions successives ou des mentions répétées d’un même chantier dans les archives.

<div style="border-left: 6px solid #2F3E46; padding: 12px; background-color:#F7FBFF; border-radius: 6px;">
<b>Résultat :</b> Ces “doublons” ne doivent pas être interprétés comme des erreurs, mais comme des répétitions documentaires typiques des pratiques de consignation de l’époque.
</div>

---

## <span style="color:#2F3E46;">4.3 Formats incorrects</span>

Les dates présentaient initialement des formats hétérogènes, mais elles ont été <b>normalisées lors du nettoyage</b>.  
Le dataset analysé ne contient donc plus de dates incohérentes ou mal formatées.

<div style="border-left: 6px solid #C97E4B; padding: 12px; background-color:#F4F1EC; border-radius: 6px;">
<b>Méthode :</b> La normalisation des dates était indispensable pour garantir une analyse temporelle fiable et reproductible.
</div>

Cette variabilité initiale reflète la diversité des sources et des modes de saisie, fréquente dans les corpus patrimoniaux.

---

## <span style="color:#2F3E46;">4.4 Incohérences patrimoniales</span>

Aucune incohérence directe n’a pu être identifiée entre les dates et les périodes, puisque la colonne <b>période</b> est vide dans le dataset d’origine.  
Cette absence empêche toute vérification croisée et limite la capacité à détecter des contradictions historiques.

<div style="border-left: 6px solid #8A5A44; padding: 12px; background-color:#F4F1EC; border-radius: 6px;">
<b>Interprétation :</b> Ce manque souligne une perte d’information patrimoniale : sans période structurée, il est difficile d’établir une chronologie précise ou de comparer les interventions entre elles.
</div>

La reconstruction de ce champ constitue donc un enjeu majeur pour la valorisation scientifique du corpus.

<div style="border-left: 6px solid #8A5A44; padding: 12px; background-color:#F4F1EC; border-radius: 6px;">

### Fusion et structuration des données patrimoniales

Les données issues du fichier CSV (voir Chapitre 2) ont été nettoyées, normalisées et divisées en deux fichiers distincts :

- <code>descriptifs_chantiers_clean.csv</code>  
- <code>localisation_clean.csv</code>

Ces deux fichiers ont ensuite été intégrés dans une base PostgreSQL nommée <b>PARIS_ARCHEO</b>.

La base comporte deux tables principales :

- <code>localisation</code>  
- <code>descriptifs_chantiers</code>

Une vue principale, <code>v_chantiers_geo</code>, combine les informations descriptives (responsable, date, nature, synthèse…) et géographiques (latitude, longitude, adresse).

<div style="border-left: 6px solid #2F3E46; padding: 12px; background-color:#F7FBFF; border-radius: 6px;">
<b>Résultat :</b> Cette structuration permet une exploration thématique riche et reproductible, adaptée aux besoins des chercheurs comme du grand public.

````python
-- SQL
-- Ce QUERY permet de créer la colonne géométrique dans la table "localisation".

ALTER TABLE localisation
ADD COLUMN geom geometry(Point, 4326);

UPDATE localisation
SET geom = ST_SetSRID(ST_MakePoint(longitude, latitude), 4326);
````

````python
-- SQL
-- Ce QUERY crée l'index spatial.

UPDATE localisation
SET geom = ST_SetSRID(ST_MakePoint(longitude, latitude), 4326);
````

````python
-- SQL

CREATE VIEW v_chantiers_geo AS
SELECT
    d.id_chantier,
    d.nature_operation,
    d.responsable,
    d.date_operation,
    d.synthese,
    d.periode,
    d.adresse,
    d.coe_postal,
    l.latitude,
    l.longitude,
    l.geom
FROM descriptifs_chantiers d
JOIN localisation l USING (id_chantier);

-- Ce QUERY permet : 
-- d'accéder à toutes les informations relatives à un chantier en une seule requête.
-- de préparer tout éventuel export vers QGIS ou une API
-- de simplifier des analyses ultérieures
````

2. Analyse temporelle des opérations archéologiques

Ce type d'analyse permet d'explorer l'évolution des interventions archéologiques dans le temps.

````python
-- SQL

CREATE VIEW v_chantiers_par_annee AS
SELECT
    EXTRACT(YEAR FROM date_operation) AS annee,
    COUNT(*) AS nb_chantiers
FROM descriptifs_chantiers
GROUP BY annee
ORDER BY annee;

-- Ce QUERY permet :
-- de visualiser les périodes d'activité archéologique
-- d'identifier des tendances (pics, creux, cycles)
-- préparer des graphiques temporels
````

3. Analyse thématique : typologique des opérations

Cette analyse permet de synthétiser les différents types d'opérations (diagnostics, fouilles...).

````python
-- SQL

CREATE VIEW v_chantiers_par_type AS
SELECT
    nature_operation,
    COUNT(*) AS nb_chantiers
FROM descriptifs_chantiers
GROUP BY nature_operation
ORDER BY nb_chantiers DESC;

-- Ce QUERY permet de connaître la nature des interventions
-- de comparer les pratiques archéologiques selon les périodes
-- de créer des graphiques selon les thèmes
````

4. Analyse spatiale avec PostGIS

PostGIS permet d'utiliser les données afin de visualiser, par example, des chantiers dans un rayon donné autour d'un point de référence. Voici un exemple de QUERY permettant de voir les chantiers dans un rayon de 500 mètres autour d'un point géographique précis.

````python
-- SQL

SELECT *
FROM v_chantiers_geo
WHERE ST_DWithin(
    geom,
    ST_SetSRID(ST_MakePoint(2.3522, 48.8566), 4326),
    500
);

-- Ce QUERY permet :
-- de préparer de futures cartes interactives
-- d'étudier des zones géographiques selon leur intérêt archéologique
-- de se rendre compte de la densité archéologique dans une zone donnée de Paris
````

5. Extraction géographique : Paris intra-muros

Cette analyse-ci extrait les chantiers dans la zone administrative de Paris (ne comprends pas la Petite ni la Grande Couronne).

````python
-- SQL

CREATE VIEW v_chantiers_paris AS
SELECT *
FROM v_chantiers_geo
WHERE geom && ST_MakeEnvelope(
    2.224, 48.815,
    2.469, 48.902,
    4326
);

-- Ce QUERY permet :
-- de filtrer les données
-- préparer des cartes thématiques par arrondissement
-- isoler les interventions dans la zone administrative de Paris
````

6. Analyse de proximité

Cette analyse permet de visualiser clairement les chantiers proches les uns des autres, ce qui nous permet de détecter des clusters, d'analyser les zones d'interventions répétées ou de préparer une éventuelle visualisation de densité.

````python
-- SQL

CREATE VIEW v_chantiers_proximite AS
SELECT
    a.id_chantier AS chantier1,
    b.id_chantier AS chantier2,
    ST_Distance(a.geom::geography, b.geom::geography) AS distance_m
FROM localisation a
FROM localisation b
ON a.id_chantier < b.id_chantier
AND ST_DWithin(a.geom, b.geom, 200);
````

7. Visualisation : buffers et zones d'influence

````python
-- SQL

CREATE VIEW v_chantiers_buffer_200m AS
SELECT
    id_chantier,
    geom,
    ST_Buffer(geom::geography, 200)::geometry AS zone_200m
FROM localisation;

-- Ce QUERY permet :
-- de visualizer les zones d'impact
-- de préparer des analyses de voisinage
-- d'enrichier les cartes avec des zones tampons
````
